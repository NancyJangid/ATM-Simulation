```python

class BankAccount:
    def __init__(self,name,account_number,balance):
        self.name = name
        self.account_number = account_number
        self.balance = balance
        self.transactions = []
    
    def account_details(self):
        print("Name: ", self.name)
        print("Account_Number: ", self.account_number)
        print("Balance: $", self.balance)
        print("_______________________________________")
        print(" ")
        print(" ")
        print(" ")

    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance = self.balance - amount
            print("Withdrawl Successful!")
            self.transactions.append({"amount":amount, "type":"debit"})
            print("Updated Balance: $", self.balance)
        else:
            print("Cannot withraw as your balance is: $", self.balance)
        print("_______________________________________")
        print(" ")
        print(" ")
        print(" ")

    def deposit(self, amount):
        self.balance = self.balance + amount
        print("Deposit Successful!")
        print("Updated Balance: $", self.balance)
        self.transactions.append({"amount":amount, "type":"credit"})
        print("_______________________________________")
        print(" ")
        print(" ")
        print(" ")

    def transfer(self, account, amount):
        if self.balance >= amount:
            self.balance = self.balance - amount
            print("Transfer Successful to account number" , account)
            print("Updated Balance: $", self.balance)
            self.transactions.append({"amount":amount, "type":"debit"})
        else:
            print("Cannot transfer as your balance is: $", self.balance)
        print("_______________________________________")
        print(" ")
        print(" ")
        print(" ")
    
    def history(self):
        for i in self.transactions:
            amount = i["amount"]
            types = i["type"]
            if types == "debit":
                print("Debit: $", amount)
            elif types == "credit":
                print("Credit: $", amount)
        print("Balance: $", self.balance)
        print("_______________________________________")
        print(" ")
        print(" ")
        print(" ")

            
        
        
        
```
